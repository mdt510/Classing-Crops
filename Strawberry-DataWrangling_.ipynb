```
import pandas as pd
from PIL import Image
import numpy as np
import glob
```

Load each image from a directory and build a list of each image as a one-dimensional Numpy array.

```
path = glob.glob('/content/drive/MyDrive/Strawberry_Grading/Strawberry Images/A class/*.jpg')

img_list=[]  #instantiate the list of image data

for i in path:
  img = Image.open(i)  #open image

  np_img = np.array(img).flatten()  #build a flattened numpy array of image

  img_list.append(np_img) #append array to df
```

Create a Pandas df from list of image data:
*   row - each image array
*   column - RGB chanel value for each pixel



```
df=pd.DataFrame(img_list)
```

save df as a .CSV for later use

```
df.to_csv('/content/drive/MyDrive/Strawberry_Grading/pixel_df.csv')
```

Create a Numpy array for each image using tensorflow:



Numpy array of size (n x m) 
*   n - number of observations:  3 (Red, Green, Blue)
*   m - number of pixels:  10,000 (100x100)

```
from tensorflow.keras.preprocessing import image

img_mat=[]

for i in path:
  img = Image.open(i)
  
  img_ts = image.img_to_array(img)  #convert image to a (3 X 10K) Numpy matrix

  img_ts = [img_ts.flatten()]  #turn matrix into a vector (1D array)

  img_mat.append(img_ts) #concatenate all of the image vectors
```

Build pandas df from list of image data - each row is an image

```
df_mat=pd.DataFrame(img_mat)
```
